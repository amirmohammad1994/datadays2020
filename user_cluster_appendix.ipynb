```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
df_user_feature = pd.read_csv('user_like.csv')
```

```python
df_user_feature.info()
```

```python
df_user_feature.head(10)
```

```python
X = df_user_feature.iloc[:, 1:53].values
X
```

```python
kmeans = KMeans(n_clusters=20, init='k-means++', max_iter=500, n_init=10)
pred_y = kmeans.fit_predict(X)
```

```python
df_user_feature['Cluster'] = pred_y
```

```python
df_user_feature
```

```python
df_user_feature.hist(column='Cluster',bins=30, figsize=(15, 10))
```

```python
df_user_cluster = df_user_feature.filter(['userId','Cluster'], axis=1)
```

```python
df_user_cluster = df_user_cluster.set_index('userId')
```

```python
df_user_cluster.head(10)
```

```python
del(df_user_feature)
```

```python
df_user_cluster.to_csv('user_cluster.csv')
```
